<span style  = "color:rosybrown"> Personalized nutrition recommendation engine </span>
A nutrition recommendation engine system is a tool that helps users get personalized recommendations based on their dietary preferences, nutritional needs, and meal plans and we will be doing this by using a known DataStructure-Binary Search Tree(BST)

<span style ="color:rosybrown"> About the DataStructure  BST(Binary search tree): </span>
A **Binary Search Tree** is a type of data structure that organises elements in such a way that it allows operations like insertion, deletion,traversal(inorder traversal) and searching with the data having no duplicate values.
<ol>
    <li>Binary search tree has atmost 2 children.</li>
    <li>All the elements of left subtree are (lESS THAN ROOT NODE)</li>
    <li>All the elememts of right subtree are (GREATER THAN ROOT NODE)</li>
</ol>
<br>
We are here going to create a balanced Binary tree with the attributes that will be mentioned further.As we have seen what were the basic Operations that could be done, their <strong>Time Complexity</strong> expalins us the Efficiency of those respective Algorithms with large data.
<ul>
    <li><b>Insertion:O(logn)</li>
    <li><b>Deletion:O(logn)</li>
    <li><b>Search:O(logn)</li>
    <li><b>Traversal:O(n)</li>
</ul>

<span style = "color:rosybrown"> **Why BST over any other Data structure**  </span>
<ol>
    <li> BST allows efficient searching of dietary profiles based on factors like calories,proteins etc </li>
    <li> In BST the search operation takes O(logn) time making it faster than linear search O(n) in arrays or linked list</li>

<span style ="color:rosybrown">class **BSTNode** </span>
This class represents a **node** in BST.A BST node typically contains a data element and points out to left node or right node.

Attributes in this class are:
<ul type="square">
    <li>Profile:This attribute is used for storing or holding the DietaryProfile</li><br>Based on a key for reference ,here we are taking calories intake as a key to start with insertion into BST.
</ul>

<ul>
    <li><strong>left:</strong>Points to smaller value node</li> 
    <li><strong>right:</strong>points to Larger Value node</li>

**loading the profiles into the BST(insertion):**



<span style="color:rosybrown">class **DietaryBST:**</span>

This **DietaryBST** class represents the binary search tree that will store instances of BSTNode,each of which holds a dietary profile. This class contains methods for inserting new nodes and searching for nodes based on person calorie_needs(**key**).

<span style = "color:rosybrown"> Methods that we are using in this are: </span>

Constructor **`__init__`** method used for  initializing an empty BST

We are using a function called **`insert`**: It inserts a new DietaryProfile into the tree, if the tree is empty a new BSTNode is created if the tree is not empty _insert method is called **recursively** to find the correct position for new node in the tree. (**`_insert`**) is a **protected** attribute i.e..the variable inside this class should not be accessed directly from outside the class but it can still be accessed if needed.

We are using a another function called **`search`** : (_search) based on calorie_needs this method starts the search for profiles using BST.








<span style = "color:rosybrown"> **---->Implemention of nutritional requirement matching** </span>

**find_best_match** method : This method performs an inorder traversal of the tree to find the profile whose calorie_needs is closest to the provided target using min()and absolute difference functions.
<br><br>
We are having this case  when the exact match is not found,for that we modify our **BST Search function** that would return the closest match if the exact value(based on key) was not found which uses **Recursion** to Traverse the BST efficiently.
